## 1.Selection 
The selection interaction method allows users to choose specific data points or regions of interest within a visualization. In the context of a scatter plot, for example, users can click on individual points to select them. This interaction enables users to focus on particular data points or subsets of data, facilitating detailed analysis or comparison. In the provided code example, users can click on data points in the scatter plot representing countries to select them, and their corresponding GDP and Population values will be displayed.

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_with_selection(data):
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    
    ax.scatter(data['GDP'], data['Population'], data['GDP Per Capita'], alpha=0.5)
    ax.set_xlabel('GDP')
    ax.set_ylabel('Population')
    ax.set_zlabel('GDP Per Capita')
    ax.set_title('3D Scatter Plot: GDP vs. Population vs. GDP Per Capita')
    
    def on_select(event):
        if event.inaxes is None:
            return
        selected_point = event.artist
        x_data = selected_point.get_xdata()
        y_data = selected_point.get_ydata()
        z_data = selected_point.get_zdata()
        ind = event.ind
        print(f"Selected data point: GDP={x_data[ind]}, Population={y_data[ind]}, GDP Per Capita={z_data[ind]}")
    
    fig.canvas.mpl_connect('pick_event', on_select)
    plt.show()

plot_with_selection(df)
```

## 2.Exploration
The exploration interaction method enables users to zoom, pan, or navigate through the data to explore it in more detail. In the provided code example, users can interactively zoom in or out of the scatter plot by using the toolbar provided by Matplotlib. They can also pan across the plot to focus on specific regions of interest. This interaction empowers users to visually inspect different parts of the dataset, facilitating a deeper understanding of the relationship between GDP and Population in this case.

```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_with_exploration(data):
    G = nx.Graph()
    

    subset_data = data.sample(min(50, len(data)))
    
    for idx, row in subset_data.iterrows():
        G.add_node(row['Country Name'])
    
    for i in range(len(subset_data)):
        for j in range(i + 1, len(subset_data)):
            source_country = subset_data.iloc[i]['Country Name']
            target_country = subset_data.iloc[j]['Country Name']
            gdp_diff = abs(subset_data.iloc[i]['GDP'] - subset_data.iloc[j]['GDP'])
            population_diff = abs(subset_data.iloc[i]['Population'] - subset_data.iloc[j]['Population'])
            edge_weight = 1 / (gdp_diff + population_diff + 1) 
            G.add_edge(source_country, target_country, weight=edge_weight)
    
    # Draw the network (Arc Diagram)
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, iterations=50)
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='red', font_size=8, font_weight='bold', alpha=0.8)

    plt.title('Arc Diagram: GDP vs. Population')
    plt.axis('off')
    plt.show()

plot_with_exploration(df)
```

The reconfigure interaction method enables users to dynamically change the visualization type or adjust parameters to customize their view of the data. ## 4.Encode
The encode interaction method involves representing data attributes using visual properties like color, size, or shape. In the provided code example, we encode the 'Continent Name' attribute using different colors in a scatter plot. Each continent is assigned a unique color, allowing users to visually distinguish between data points belonging to different continents. This encoding enhances the visualization by adding an additional dimension of information (continent) using the visual property of color, facilitating easier identification and analysis of patterns or differences in the dataset across continents.

```python
continent_colors = {'Africa': 'red', 'Asia': 'blue', 'Europe': 'green', 'North America': 'orange', 'Oceania': 'purple', 'South America': 'yellow'}

def plot_with_encode(data):
    plt.figure(figsize=(10, 6))
    
    continents = list(continent_colors.keys())
    num_continents = len(continents)
    gdp_data = [data[data['Continent Name'] == continent]['GDP'].mean() for continent in continents]
    population_data = [data[data['Continent Name'] == continent]['Population'].mean() for continent in continents]
    
    bar_width = 0.35
    index = np.arange(num_continents)
    
    plt.bar(index, gdp_data, bar_width, color='blue', label='GDP')
    
    plt.bar(index + bar_width, population_data, bar_width, color='green', label='Population')
    
    plt.xlabel('Continent')
    plt.ylabel('Average Value')
    plt.title('Average GDP and Population by Continent')
    plt.xticks(index + bar_width / 2, continents)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()

plot_with_encode(df)
```

## 5.Abstract
The abstract interaction method involves providing summary statistics or aggregate views to help users understand the data at different levels of granularity. In the provided code example, we use histograms to visualize the distribution of GDP and Population attributes. Histograms offer a summarized view of the data's distribution, showing the frequency of values within predefined intervals (bins). By observing the histograms, users can quickly grasp the spread and central tendency of these attributes, gaining insights into the dataset's overall characteristics. This abstract representation facilitates a high-level understanding of the data's distribution and aids users in identifying potential patterns or outliers.

```python
def plot_with_abstract(data):
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.hist(data['GDP'], bins=20, color='skyblue', edgecolor='black')
    plt.xlabel('GDP')
    plt.ylabel('Frequency')
    plt.title('Distribution of GDP')
    
    plt.subplot(1, 2, 2)
    plt.hist(data['Population'], bins=20, color='salmon', edgecolor='black')
    plt.xlabel('Population')
    plt.ylabel('Frequency')
    plt.title('Distribution of Population')
    
    plt.tight_layout()
    plt.show()

plot_with_abstract(df)
```

## 6.Elaborating  
The elaborate interaction method enriches the visualization by offering tooltips or additional information on demand when users hover over data points. In the provided code example, tooltips are displayed with details about countries, including their education expenditure (% GDP) and inflation rate, when hovering over data points in a scatter plot. This interactive feature allows users to gain contextual insights into individual data points, enhancing their understanding of the dataset and supporting more informed analysis and decision-making.

```python
def plot_with_elaborate(data):
    # Group the data by 'Year' and calculate the mean values
    grouped_data = data.groupby('Year').mean()

    # Create the time series plot
    plt.figure(figsize=(10, 6))
    plt.plot(grouped_data.index, grouped_data['Education Expenditure (% GDP)'], marker='o', label='Education Expenditure (% GDP)')
    plt.plot(grouped_data.index, grouped_data['Inflation Rate'], marker='o', label='Inflation Rate')
    
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title('Time Series: Education Expenditure vs. Inflation Rate')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_with_elaborate(df)
```

## 7.Filter
The filter interaction method enables users to dynamically restrict the data displayed in the visualization based on criteria they specify. In the provided code example, users can filter countries by their Education Expenditure (% GDP) values using a slider widget. By adjusting the slider, users can focus their analysis on specific subsets of the data, facilitating targeted exploration and insights into the relationship between Unemployment and Inflation Rate under different education expenditure levels.

```python
import ipywidgets as widgets
from IPython.display import display

def update_plot(threshold):
    plt.figure(figsize=(10, 6))
    filtered_data = df[df['Education Expenditure (% GDP)'] >= threshold]
    plt.scatter(filtered_data['Unemployment'], filtered_data['Inflation Rate'], alpha=0.5)
    plt.xlabel('Unemployment')
    plt.ylabel('Inflation Rate')
    plt.title(f'Scatter Plot: Unemployment vs. Inflation Rate (Education Expenditure >= {threshold}%)')
    plt.grid(True)
    plt.show()

threshold_slider = widgets.FloatSlider(
    min=0,
    max=20,
    step=0.5,
    value=5,
    description='Education Expenditure Threshold (% GDP):',
    continuous_update=False
)

def slider_eventhandler(change):
    update_plot(change.new)

threshold_slider.observe(slider_eventhandler, names='value')

display(threshold_slider)

update_plot(5)
```

## 8.Connect 
The connect interaction method allows users to establish connections or relationships between different data points or entities. In the provided code example, we create a network graph to visualize connections between countries based on their geographical proximity, such as belonging to the same continent. By representing countries as nodes and establishing edges between them based on their proximity, users can visually identify relationships and patterns within the dataset. This interactive feature facilitates the exploration of connections between entities, providing insights into geographical relationships and potential interactions between countries.

```python
import networkx as nx

def create_network_graph(data):
    G = nx.Graph()
    for index, row in data.iterrows():
        G.add_node(row['Country Name'])
    
    for continent in data['Continent Name'].unique():
        countries_in_continent = data[data['Continent Name'] == continent]['Country Name'].tolist()
        for country1 in countries_in_continent:
            for country2 in countries_in_continent:
                if country1 != country2:
                    G.add_edge(country1, country2)
    
    return G

def plot_network_graph(graph):
    plt.figure(figsize=(10, 8))
    nx.draw(graph, with_labels=True, node_color='skyblue', node_size=1000, font_size=10, edge_color='gray')
    plt.title('Network Graph: Countries and Their Connections')
    plt.show()

graph = create_network_graph(df)

plot_network_graph(graph)
```

## Customized Interaction with Event Handling
Customized Interaction with Event Handling involves designing interactive features in a graphical user interface (GUI) that respond to user actions, such as mouse clicks, keyboard inputs, or other events. This allows users to interact with visual elements dynamically, enabling tasks such as data exploration, selection, filtering, or manipulation. By customizing event handling, developers can create more intuitive and engaging user experiences, enhancing usability and enabling users to gain deeper insights from the data presented in the interface.

```python
x = df['Import']
y = df['Export']
colors = df['Education Expenditure (% GDP)']
sizes = df['Health Expenditure (% GDP)'] * 150 
fig, ax = plt.subplots()
sc = ax.scatter(x, y, s=sizes, c=colors, alpha=0.5)


def on_mouse_enter(event):
    print('Mouse entered the figure')

def on_mouse_leave(event):
    print('Mouse left the figure')

def on_pick(event):
    ind = event.ind
    print(f'You picked {len(ind)} point(s): {ind}')
    for i in ind:
        print(f'Country: {df.iloc[i]["Country Name"]}, Import: {df.iloc[i]["Import"]}, Export: {df.iloc[i]["Export"]}')

def on_key(event):
    if event.key == 'r':
        sc.set_sizes(df['Health Expenditure (% GDP)'] * 150)
        fig.canvas.draw()
        print('Sizes randomized')


fig.canvas.mpl_connect('figure_enter_event', on_mouse_enter)
fig.canvas.mpl_connect('figure_leave_event', on_mouse_leave)
fig.canvas.mpl_connect('pick_event', on_pick)
fig.canvas.mpl_connect('key_press_event', on_key)

plt.show()
```

## Algorithm Enhancement
Algorithm enhancement involves improving an existing method by introducing new techniques or approaches to provide additional insights or functionality. This enhancement could include introducing novel algorithms, statistical methods, or computational techniques to improve the performance, accuracy, or usability of the existing method. The goal is to enhance the functionality of the method and provide users with more valuable information or capabilities when interacting with the data.

```python
x = df['Population Density']
y = df['GDP Per Capita']

fig, ax = plt.subplots()
sc = ax.scatter(x, y, picker=5)  

def on_pick(event):
    ind = event.ind
    print(f'You picked {len(ind)} point(s): {ind}')
    for i in ind:
        country_name = df.iloc[i]["Country Name"]
        gdp_per_capita = df.iloc[i]["GDP Per Capita"]
        is_outlier = False

        Q1 = df['GDP Per Capita'].quantile(0.5)
        Q3 = df['GDP Per Capita'].quantile(1.5)
        IQR = Q3 - Q1

        if gdp_per_capita < (Q1 - 3 * IQR) or gdp_per_capita > (Q3 + 3 * IQR):
            is_outlier = True

        print(f'Country: {country_name}, GDP Per Capita: {gdp_per_capita}')
        if is_outlier:
            print('GDP per capita value is an outlier.')

fig.canvas.mpl_connect('pick_event', on_pick)

plt.show()
```

## Connecting events
Connecting events to their respective functions involves establishing a link between specific events triggered by user interactions (such as clicking, picking, key pressing, or mouse movement) and the corresponding functions that should be executed in response to those events. This linkage enables the program to respond dynamically to user actions, enhancing interactivity and usability in graphical user interfaces. By connecting events to their respective functions, developers can create more intuitive and responsive applications that provide users with a seamless interactive experience.

```python
x = df['GDP Per Capita']
y = df['Education Expenditure (% GDP)']
colors = df['Health Expenditure (% GDP)']
sizes = df['Population'] / 1e6 

fig, ax = plt.subplots()
sc = ax.scatter(x, y, s=sizes, c=colors, alpha=0.5)

def on_pick(event):
    ind = event.ind
    print(f'You picked {len(ind)} point(s): {ind}')
    for i in ind:
        country_name = df.iloc[i]["Country Name"]
        gdp_per_capita = df.iloc[i]["GDP Per Capita"]
        education_expenditure = df.iloc[i]["Education Expenditure (% GDP)"]
        health_expenditure = df.iloc[i]["Health Expenditure (% GDP)"]
        population = df.iloc[i]["Population"]
        print(f'Country: {country_name}, GDP Per Capita: {gdp_per_capita}, Education Expenditure (% GDP): {education_expenditure}, Health Expenditure (% GDP): {health_expenditure}, Population: {population}')

def on_click(event):
    if event.inaxes == ax:
        print(f'Clicked at: {event.xdata}, {event.ydata}')

def on_key(event):
    if event.key == 'r':
        ax.clear()
        sc = ax.scatter(x, y, s=sizes, c=colors, alpha=0.5)
        plt.draw()
        print('Plot reset')

def on_mouse_move(event):
    if event.inaxes == ax:
        print(f'Mouse moved to: {event.xdata}, {event.ydata}')

fig.canvas.mpl_connect('pick_event', on_pick)
fig.canvas.mpl_connect('button_press_event', on_click)
fig.canvas.mpl_connect('key_press_event', on_key)
fig.canvas.mpl_connect('motion_notify_event', on_mouse_move)

plt.show()
```
